# Data visualization - Intro 

## 1. Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Read data into a dataframe

```python
df = pd.read_csv('people-example.csv')
df.head()
```

```python
df.tail()
```

## 3. Plot the data

```python
%matplotlib inline
plt.figure(); df.plot()
```

### Arguments for plot function 
#### 1. Specify x and y axis

```python
df.plot(x='Last Name', y='age' )
```

#### 2. Specify kind of plot

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
df2 = pd.read_csv('iris.csv')
df2.plot.hist(x = 'Name')
```

```python
df2.head()
```

```python
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(df2, 'Name')
```

```python
from pandas.tools.plotting import andrews_curves
andrews_curves(df2,'Name')
```
